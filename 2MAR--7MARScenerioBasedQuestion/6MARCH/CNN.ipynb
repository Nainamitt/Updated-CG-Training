```python
# Scenario Question
# A machine learning engineer is designing a convolutional neural network (CNN) to process grayscale images of handwritten digits
#  (28×28 pixels). She starts with a single convolutional layer defined as follows
```

```python
# 1 method using torch
```

```python
import torch
import torch.nn as nn
```

```python
conv = nn.Conv2d(
    in_channels  = 1,    # grayscale input
    out_channels = 32,   # 32 filters -> 32 feature maps
    kernel_size  = 3,    # 3x3 filter
    stride       = 1,
    padding      = 1     # SAME padding: output = same size
)
```

```python
x   = torch.randn(1, 1, 28, 28) # (batch, C, H, W)
out = conv(x)
print(out.shape) # torch.Size([1, 32, 28, 28])
```

```python
print('Weights:', conv.weight.shape) # (32, 1, 3, 3)
print('Bias   :', conv.bias.shape)   # (32,)
print('Total  :', 32*1*3*3 + 32)     # = 320
```

```python
# 2 Method using tensorflow
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
conv = keras.layers.Conv2D(
    filters     = 32,
    kernel_size = (3, 3),
    strides     = (1, 1),
    padding     = 'same',
    activation  = 'relu',
    input_shape = (28, 28, 1)
)
```

```python
# Build and summarise
model = keras.Sequential([conv])
model.build(input_shape=(None, 28, 28, 1))
model.summary()
# Output shape: (None, 28, 28, 32)
# Trainable params: 320
```

```python
# Scenario Question

# A biomedical researcher is designing a convolutional neural network (CNN) to analyze microscopic grayscale images of blood cells (64×64 pixels). She begins with a single convolutional layer defined as follows:

# Input: 64×64 grayscale images (1 channel)
# Convolutional Layer: 16 filters, each of size 5×5
# Stride: 1
# Padding: Same
# Activation Function: ReLU
# The goal of this layer is to extract local structural features such as cell boundaries and texture variations, which will later help in classifying whether the blood cells are healthy or show signs of abnormality.
```

```python
import torch
import torch.nn as nn

conv = nn.Conv2d(
    in_channels=1,
    out_channels=16,
    kernel_size=5,
    stride=1,
    padding=2   # same padding for 5x5
)

relu = nn.ReLU()

x = torch.randn(1, 1, 64, 64)

out = relu(conv(x))

print(out.shape)
```

```python
print("Weights shape:", conv.weight.shape)
print("Bias shape:", conv.bias.shape)
print("Total params:", 5*5*1*16 + 16)
```

```python
import tensorflow as tf
from tensorflow.keras import layers

# Create convolution layer
conv = layers.Conv2D(
    filters=16,
    kernel_size=(5,5),
    strides=1,
    padding='same',
    activation='relu'
)

# Example input image (batch size = 1)
x = tf.random.normal((1, 64, 64, 1))

# Pass input through convolution layer
output = conv(x)

print("Output shape:", output.shape)
```
