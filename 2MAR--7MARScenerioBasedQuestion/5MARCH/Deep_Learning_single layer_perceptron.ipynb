```python
# Scenario:
# We want to classify whether a restaurant order is “Large” (1) or “Small” (0) based on the number
#  of items ordered.
# - If items ≥ 3 → Large order
# - If items < 3 → Small order
```

```python
import numpy as np

# Input: number of items
X = np.array([[1], [2], [3], [4], [5]])

# Output: order size
y = np.array([0, 0, 1, 1, 1])
```

```python
weights = np.random.rand(1)
bias = np.random.rand(1)

print(weights)
print(bias)


learning_rate = 0.1
```

```python
def activation(z):
    return 1 if z >= 0 else 0
```

```python
for epoch in range(50):
    print("Epoch:", epoch+1)

    for i in range(len(X)):

        # Linear equation
        z = np.dot(X[i], weights) + bias

        # Prediction
        y_pred = activation(z)

        # Error
        error = y[i] - y_pred

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(X[i], y_pred, y[i], error)
```

```python
test_orders = np.array([1,2,3,5])

for order in test_orders:
    z = np.dot(order, weights) + bias
    print("Items:", order, "Prediction:", activation(z))
```
