```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

article = """
Artificial Intelligence (AI) is rapidly transforming the field of education.
AI-powered systems can analyze large amounts of student performance data
and recommend personalized learning paths for students. Intelligent tutoring
systems provide real-time feedback and help students understand difficult
concepts. Universities are also using AI for administrative tasks such as
automated grading, academic advising, and course scheduling. However,
challenges such as data privacy, algorithmic bias, and ethical concerns
must be addressed before AI can be widely adopted in education systems.
"""

prompt = "Summarize the following research article:\n" + article + "\nSummary:"

result = generator(prompt, max_length=120, num_return_sequences=1)

# Print generated summary
print("Generated Summary:\n")
print(result[0]['generated_text'])
```

```python
# Task 2 – Prompt Engineering Experiments 
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

article = """
Artificial Intelligence (AI) is rapidly transforming education.
AI-powered tools analyze student performance and create personalized learning paths.
Intelligent tutoring systems provide real-time feedback to help students understand difficult topics.
Universities are also using AI to automate administrative tasks such as grading and advising.
However, issues such as data privacy and ethical concerns must be addressed.
"""
```

```python
# 1️⃣ Zero-Shot Prompt
# Ask the model to summarize the article without providing examples.

# Example task:
# Summarize the following article in 5 bullet points.
```

```python
zero_prompt = "Summarize the following article in 5 bullet points:\n\n" + article + "\nSummary:"
result = generator(zero_prompt, max_length=120, num_return_sequences=1)

print("Zero Shot Output:\n")
print(result[0]["generated_text"])
```

```python
# 2️⃣ Few-Shot Prompt
# Provide 2 example summaries before asking the model to summarize a new article.
```

```python
few_prompt = """
Example 1
Article: Solar energy helps reduce pollution and provides renewable power.
Summary:
• Solar energy is a renewable power source.
• It reduces environmental pollution.
• It supports sustainable energy production.

Example 2
Article: Machine learning helps doctors detect diseases earlier using patient data.
Summary:
• Machine learning analyzes medical data.
• It helps detect diseases early.
• Doctors use AI tools for diagnosis.

Now summarize the following article in bullet points:

""" + article + "\nSummary:"

result = generator(few_prompt, max_length=150, num_return_sequences=1)

print("Few Shot Output:\n")
print(result[0]["generated_text"])
```

```python
# 3️⃣ Chain-of-Thought Prompt
# Ask the model to reason step-by-step before generating the summary.

# Example structure:

# Analyze the article step by step:

# 1. Identify the main topic

# 2. Extract key ideas

# 3. Generate a concise summar
```

```python
cot_prompt = """
Analyze the article step by step:

1. Identify the main topic
2. Extract key ideas
3. Generate a concise summary

Article:
""" + article + """

Answer:
"""

result = generator(cot_prompt, max_length=180, num_return_sequences=1)

print("Chain of Thought Output:\n")
print(result[0]["generated_text"])
```

```python
# Task 3 – Prompt Optimization
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

optimized_prompt = """
You are an academic research assistant.

Read the article and generate an executive summary in a professional tone.

Output format:

Key Insights:
1.
2.
3.

Actionable Takeaway:
1 recommendation based on the article.

Article:
"""
```

```python
article1 = """
Artificial Intelligence is transforming healthcare by enabling faster diagnosis
and personalized treatment. AI systems analyze medical images and patient data
to detect diseases such as cancer earlier than traditional methods.
Hospitals are increasingly using AI tools to assist doctors in clinical
decision-making and patient monitoring. However, challenges related to
data privacy, transparency, and reliability must still be addressed.
"""

prompt1 = optimized_prompt + article1

result1 = generator(prompt1, max_length=180, num_return_sequences=1)

print("Output for Article 1:\n")
print(result1[0]["generated_text"])
```

```python
article2 = """
Blockchain technology provides a decentralized and secure system
for recording digital transactions. It is widely known for supporting
cryptocurrencies such as Bitcoin, but it is also used in supply chain
management, healthcare data storage, and digital identity verification.
Blockchain improves transparency, security, and trust between
organizations and users.
"""

prompt2 = optimized_prompt + article2

result2 = generator(prompt2, max_length=180, num_return_sequences=1)

print("\nOutput for Article 2:\n")
print(result2[0]["generated_text"])
```

```python
# Task 4 – Tokenization Experiment 
```

```python
from transformers import AutoTokenizer

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Sample paragraph
text = """
Artificial Intelligence is transforming education by enabling personalized learning experiences.
AI systems analyze student performance data and provide customized learning paths.
"""

# Tokenize the text
tokens = tokenizer.tokenize(text)

# Count tokens
token_count = len(tokens)

# Display results
print("Tokenized Output:\n")
print(tokens)

print("\nTotal Token Count:")
print(token_count)
```

```python
# Task 5 – Build a Mini AI Tool
```

```python
from transformers import pipeline

# Load LLM model
generator = pipeline("text-generation", model="gpt2")

# Accept article input from user
article = input("Enter a long article:\n")

# Prompt for structured output
prompt = """
You are an AI research assistant.

Read the article and generate:

Short Summary:
Explain the article briefly.

Key Insights:
1.
2.
3.

Actionable Recommendation:
Provide one practical recommendation based on the article.

Article:
""" + article + """

Answer:
"""

# Generate response
result = generator(prompt, max_length=200, num_return_sequences=1)

# Display structured output
print("\nAI Research Assistant Output\n")
print(result[0]["generated_text"])
```
